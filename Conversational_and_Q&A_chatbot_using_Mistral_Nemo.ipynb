```python
import torch
from langchain.chains import RetrievalQA
from langchain.llms import HuggingFaceHub
import os
from google.colab import userdata
import requests
import transformers
from langchain.embeddings import HuggingFaceEmbeddings
from huggingface_hub import InferenceClient
from transformers import AutoTokenizer, AutoModelForSequenceClassification

```

```python
def generate_response_from_inference_api(user_question, conversation_history):
    """Sends a request to the Hugging Face Inference API."""
    sec_key = userdata.get("HUGGINGFACEHUB_API_TOKEN")
    os.environ["HUGGINGFACEHUB_API_TOKEN"] = sec_key

    # Build the prompt with conversation history
    prompt = f"""
    You are a helpful and friendly chatbot.
    Conversation:
    {conversation_history}
    Human: {user_input}
    Chatbot: generate only one output as a string
    """

    client = InferenceClient("mistralai/Mistral-Nemo-Instruct-2407", token=sec_key)
    response = client.text_generation(prompt, max_new_tokens=500)
    return response

conversation_history = []
```

```python
# Example usage

while True:
    user_input = input("You: ")
    conversation_history.append(f"You: {user_input}")

    if user_input.lower() == "exit":
        break

    response = generate_response_from_inference_api(user_input,conversation_history)
    conversation_history.append(f"Bot: {response}")
    print(f"Bot: {response}")
```
